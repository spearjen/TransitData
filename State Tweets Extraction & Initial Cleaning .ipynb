```python
from config import api_key, api_secret, bearer_token, access_token, access_secret
import tweepy
import time
#This sets the authorization for tweepy, seems to work
#Have to set auth every time. 
auth = tweepy.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)
```

```python
#List of states to build tweet dictionary
states = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
state_tweets = dict()
```

```python
print(len(state_tweets))
```

```python
#Try building state dictionary here
loop = 0
for i in states:
    for status in tweepy.Cursor(api.search, q = f'public transit {i} exclude:retweets', tweet_mode = 'extended').items(100000):
        try:
            state_tweets[i] = [status.full_text]
            time.sleep(1.5)
        except UnicodeEncodeError:
            print("Unable to process this tweet, moving on.....")
            pass
        loop +=1
        print(f"Loop {loop} complete.")
        print(f"State {i} complete.")
print(state_tweets)
```

#### Now, let's work on cleaning data. 

```python
#Ex of what a sample tweet text looks like.
state_tweets['Tweets'].loc[3]
```

```python
#Now, some potential regex for cleaning this data. 
def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
#Try applying to the column, see what happens.
tweets_clean = pd.DataFrame(state_tweets.Tweets.apply(round1))
tweets_clean
```

```python
##Second round of cleaning
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‚Äò‚Äô‚Äú‚Äù‚Ä¶]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
tweets_clean = pd.DataFrame(tweets_clean.Tweets.apply(round2))
tweets_clean
```

```python
#Re-append states to the tweets
state_labels = state_tweets['State']
state_labels.head()
frames = [tweets_clean, state_labels]
```

```python
#Re-append states to the tweets
state_tweets_clean = pd.concat(frames, axis = 1)
state_tweets_clean.head()
```

Use CountVectorizer to take out stop words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english')
tweets_cv = cv.fit_transform(state_tweets_clean.Tweets)
tweets_cv
```

```python
tweets_dtm = pd.DataFrame(tweets_cv.toarray(), columns = cv.get_feature_names())
tweets_dtm.index = state_tweets_clean.State
tweets_dtm
```

```python
##Pickle the DTM just in case 
tweets_dtm.to_pickle("dtm.pkl")
```
