```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read in corpus
corpus = pd.read_csv('state_tweets_corpus.csv')
corpus.head()
```

```python
#Create lambda fxns for polarity and subjectivity
from textblob import TextBlob

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

corpus['polarity'] = corpus['tweets'].apply(pol)
corpus['subjectivity'] = corpus['tweets'].apply(sub)
corpus
```

```python
sentiment = corpus[['state', 'polarity', 'subjectivity']]
```

```python
sent_alabama = sentiment.loc[0]
```

```python
sent_alabama = sent_alabama.to_frame()
```

```python
sent_alabama.columns = ['sentiment']
```

```python
pol = sent_alabama.loc['polarity'][0]
subj = sent_alabama.loc['subjectivity'][0]
labels = ['polarity, subjectivity']
heights = [pol, subj]
```

```python
plt.bar(labels, heights)
plt.show()
```
