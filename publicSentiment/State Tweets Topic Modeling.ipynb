```python
#Dependencies
import pandas as pd
from gensim import matutils, models
import scipy.sparse
```

```python
#Bring in document-term matrix
dtm = pd.read_csv('state_tweets_dtm.csv')
dtm.head()
```

```python
##Transpose data to a term-document matrix to fit w/gensim specs
tdm = dtm.transpose()
tdm.head()
```

```python
##Transform data from df-->sparse matrix-->gensim corpus
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```
